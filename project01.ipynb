```python
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.api as sm
import os
```

```python
df = pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")
```

```python
df
```

# Borough with the maximum number of crashes reported since 2012

```python
df_2012 = df[pd.to_datetime(df['CRASH DATE_CRASH TIME']) >= pd.to_datetime('2012')]
df_2012
```

```python
borough_crash_counts = df_2012['BOROUGH'].value_counts()
borough_crash_counts
```

```python
highest_crash_borough = borough_crash_counts.idxmax()
highest_crash_borough
```

# Borough with the maximum number of crashes adjusted for population

```python
population = {
    'BOROUGH': ['BRONX', 'BROOKLYN', 'MANHATTAN', 'QUEENS', 'STATEN ISLAND'],
    'population': [1446788, 2648452, 1638281, 2330295, 487155]
}
population_df = pd.DataFrame(population)
population_df
```

```python
borough_crash_counts
```

```python
combined_df = pd.merge(population_df, borough_crash_counts, left_on='BOROUGH', right_index=True)
combined_df
```

```python
combined_df['crashes per 100k'] = (combined_df['count'] / combined_df['population']) * 100000
combined_df
```

```python
highest_borough_100k = combined_df['BOROUGH'][combined_df['crashes per 100k'].idxmax()]
highest_borough_100k
```

```python
highest_crash_100k = combined_df['crashes per 100k'].max()
highest_crash_100k
```

# Analyzing the leading cause of crashes

```python
df2 = df[df['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']
df2
```

```python
causes = df2['CONTRIBUTING FACTOR VEHICLE 1'].value_counts()
```

```python
proportion_leading = causes.max() / len(df2)
proportion_leading
```

# Top 3 causes of crashes

```python
causes.head(3)
```

```python
proportion_top3 = causes.head(3).sum() / len(df2)
proportion_top3
```
